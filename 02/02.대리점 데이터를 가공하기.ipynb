### 금액의 결측치를 수정하자

```python
uriage['item_price'].isna().sum()
```

```python
import numpy as np
```

```python
flg_is_null = uriage['item_price'].isna()
```

```python
for trg in list(uriage.loc[flg_is_null,'item_name'].unique()) :
    price = uriage.loc[(~flg_is_null) & (uriage['item_name']== trg ),'item_price'].max()
    uriage['item_price'].loc[(flg_is_null) & (uriage['item_name'] == trg)] = price
uriage.head()
```

```python
uriage['item_price'].isna().sum()
```

### 날짜 오류를 수정해보자

```python
fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_seiral,"등록일"].astype("float"),unit = "D") + pd.to_datetime("1900/01/01")
```
